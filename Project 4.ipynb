```python
import matplotlib.pyplot as plt

# Filter the data
#dt = data.head(20)
Age_data = data['Age'].head(10)
TimeOfVisit_data = data['TimeOfVisit'].head(10)

# Plotting the pie chart
plt.pie(Age_data, labels=TimeOfVisit_data, autopct='%1.1f%%', startangle=140)
plt.title("Pie chart for weekday  and temp	")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = data['PreferredCuisine'].head(5)
sizes= data['FoodRating'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart For PreferredCuisine and FoodRating")
plt.show()
```
