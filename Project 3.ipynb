```python
import matplotlib.pyplot as plt

# Filter the data
dt = data.head(20)
temp_data = dt.loc[dt['weekday'] >=0, 'temp']
weekday_data = dt.loc[dt['weekday'] >= 0, 'weekday']

# Plotting the pie chart
plt.pie(temp_data, labels=weekday_data, autopct='%1.1f%%', startangle=140)
plt.title("Pie chart for weekday  and temp	")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = data['temp'].head(5)
sizes= data['atemp'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart For temp and atemp")
plt.show()
```

```python
dt=data[data.index.isin(range(2,100))]
WD = dt['workingday'].head(70)
WE = dt['weekday'].head(70)

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(WD, WE)
plt.subplot(132)
plt.scatter(WD, WE)
plt.subplot(133)
plt.plot(WD, WE)
plt.suptitle('Categorical Plotting btw Weekdays and Workingdays')
plt.show()
```
