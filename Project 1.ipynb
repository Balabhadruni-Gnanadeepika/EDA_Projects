```python
import matplotlib.pyplot as plt

# Filter the data
dt=data[data.index.isin(range(2,200))]
Players_data = dt.loc[dt['Runs'] > 200, 'Player']
Runs_data = dt.loc[dt['Runs'] > 200, 'Runs']  # Ensure labels match the filtered data

# Plotting the pie chart
plt.pie(Runs_data, labels=Players_data, autopct='%1.1f%%', startangle=90)
plt.title("Players Which has more Runs")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data
NO_data = data.loc[data['NO'] > 2, 'NO']
Team_data = data.loc[data['NO'] > 2, 'Team']  # Ensure labels match the filtered data

# Plotting the pie chart
plt.pie(NO_data, labels=Team_data, autopct='%1.1f%%', startangle=140)
plt.title("Teams and their NOT OUT chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data
dt=data[data.index.isin(range(2,100))]
Players_data = dt.loc[dt['Inns'] > 7, 'Player']
Inns_data = dt.loc[dt['Inns'] > 7, 'Inns']  # Ensure labels match the filtered data

# Plotting the pie chart
plt.pie(Inns_data, labels=Players_data, autopct='%1.1f%%', startangle=90)
plt.title("Players and their Inns")
plt.show()
```

```python
dt=data[data.index.isin(range(2,5))]
Teams = dt['Team'].head(4)
Runss = dt['Runs'].head(4)

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(Teams, Runss)
plt.subplot(132)
plt.scatter(Teams, Runss)
plt.subplot(133)
plt.plot(Teams, Runss)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def rel_x_y(data4,x,y):
    sns.jointplot(data=data4,x=x,y=y,hue='Inns')
    plt.show()
rel_x_y(data4, 'Mat', 'Inns')
```
